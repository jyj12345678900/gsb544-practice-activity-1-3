**1. Create side-by-side boxplots of the mpg variable by the different values of the am variable. What happens if you convert the am variable back to float values and then try to create this same plot? Explain the differences between the two plots.**




```python
from plotnine import ggplot, aes, geom_boxplot


df["am"] = df["am"].astype(str)

(ggplot(df, aes(x = "am", y = "mpg",))
+ geom_boxplot()
)
```

```python
df["am"] = df["am"].astype(float)
(ggplot(df, aes(x = "am", y = "mpg",))
+ geom_boxplot()
)
```

Explanation of differences

●Categorical (str):

1.Plotnine treats "0" and "1" as two separate groups.

2.The boxplot is drawn side by side: one box for automatic cars (0), and one box for manual cars (1).

3.The x-axis shows discrete labels (0, 1).

●Numeric (float):

1.Plotnine interprets am as a continuous variable on the x-axis.

2.Geom_boxplot() groups continuous variables into one box that summarizes the whole column, instead of splitting by unique values..



**2. Create overlaid histograms of the hp variable for the different values of the cyl variable. What happens if you convert the cyl variable back to float values and then try to create this same plot? Explain the differences between the two plots.**

```python
from plotnine import ggplot, aes, geom_histogram

df["cyl"] = df["cyl"].astype(str)

(ggplot(df, aes(x="hp", fill="cyl"))
    + geom_histogram(position="identity", alpha=0.6, bins=20)
        )
```

```python
df["cyl"] = df["cyl"].astype(float)


(ggplot(df, aes(x="hp", fill="cyl"))
    + geom_histogram(position="identity", alpha=0.6, bins=20)
        )
```

Explanation of differences


1.Numeric cyl (continuous scale): plotnine treats fill="cyl" as a continuous aesthetic. Histograms aggregate counts over hp, so there isn’t a natural way to split by a continuous fill

2.Discrete cyl (factor/string): three overlaid histograms, one each for 4/6/8 cylinders, with a discrete legend ("4", "6", "8").

```python
# Create plot with first new geometry here

from plotnine import ggplot, aes, geom_point, geom_smooth, labs

(   ggplot(df, aes(x="disp", y="hp", color="cyl"))
    + geom_point(size=5, alpha=0.6)
    + geom_smooth(method="lm", se=True, color="red")
    + labs(title="Scatter Plot: Displacement(Disp) vs Horsepower(hp) by Cylinders")
)
```

```python
# Create plot with second new geometry here

from plotnine import geom_violin, labs

df["gear"] = df["gear"].astype(str)

( ggplot(df, aes(x="gear", y="hp", fill="gear"))
+ geom_violin(alpha=0.6)
+ labs(title="Violin Plot: Horsepower Distribution by Gear")
    )

```
